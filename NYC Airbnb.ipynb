```python
import geopandas as gpd
```

```python
bor_count = df.groupby('neighbourhood_group').agg('count').reset_index()
```

```python
nyc = gpd.read_file(gpd.datasets.get_path('nybb'))
nyc.head()
```

```python
bor_count.rename(columns={'neighbourhood_group':'BoroName'}, inplace=True)
bc_geo = nyc.merge(bor_count, on='BoroName')
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,10))
bc_geo.plot(column='id', cmap='Spectral_r', alpha=0.5, ax=ax, legend=True)
bc_geo.apply(lambda x: ax.annotate(s=x.BoroName, color='black', xy=x.geometry.centroid.coords[0], ha='center'), axis=1)
plt.title('Airbnb Listings in NYC')
plt.axis('off')
```

Next we are going to use a file from https://data.cityofnewyork.us/City-Government/Neighborhood-Tabulation-Areas/cpf4-rkhq to make neighbourhood map.

```python
hood = pd.read_csv(r'D:\data\NYC_airbnb\nynta.csv')
hood.head()
```

```python
from shapely import wkt
```

```python
hood.rename(columns={'NTAName':'neighbourhood'}, inplace=True)

hood['geom'] = hood ['the_geom'].apply(wkt.loads)

hood = gpd.GeoDataFrame(hood, geometry='geom')
data = gpd.GeoDataFrame(df, geometry = gpd.points_from_xy(df.longitude, df.latitude))

joined = gpd.sjoin(hood, data, how='inner', op='intersects')

joined.head()
```

```python
joined.drop(columns='geom', inplace=True)
```

```python
joined.rename(columns={'neighbourhood_left':'neighbourhood'}, inplace=True)

join_count = joined.groupby('neighbourhood').agg('count').reset_index()
```

```python
true_count = hood.merge(join_count, on='neighbourhood')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))

base = hood.plot(color='white', edgecolor='black', ax=ax)

true_count.plot(column='id', cmap='inferno_r', ax=base, legend=True)

plt.title('Airbnb listings by Neighbourhood')
```
